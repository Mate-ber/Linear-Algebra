```python
def build_matrix(ls, vs):
    J_blocks = []
    S_columns = []
    
    # Iterate over eigenvalues and their corresponding generalized eigenvectors
    for k, (lambda_k, chain) in enumerate(zip(ls, vs)):
        alpha_k = len(chain)
        
        # Create the Jordan block for eigenvalue lambda_k
        J_k = np.zeros((alpha_k, alpha_k))
        for i in range(alpha_k):
            J_k[i, i] = lambda_k
            if i + 1 < alpha_k:
                J_k[i, i + 1] = 1
        
        # Add the Jordan block to the list
        J_blocks.append(J_k)
        
        # Add the generalized eigenvectors for this eigenvalue to the eigenvector matrix S
        for vec in chain:
            S_columns.append(vec)
    
    # Combine all Jordan blocks into the Jordan matrix J
    J = np.block([[block if i == j else np.zeros_like(block) for j, block in enumerate(J_blocks)] for i in range(len(J_blocks))])
    
    # Combine all eigenvectors into the matrix S
    S = np.column_stack(S_columns)
    
    # Ensure that S is square before attempting to compute S^-1
    if len(S[0]) != len(S):
        raise ValueError(f"Matrix S must be square, but got shape {S.shape}. Please check the input eigenvectors.")
    
    # Calculate matrix A = S J S^-1
    try:
        S_inv = np.linalg.inv(S)
        A = S @ J @ S_inv
    except np.linalg.LinAlgError as e:
        raise ValueError(f"Error inverting matrix S: {e}")
    
    return J, S, A

 
```

```python
inputs = json_tricks.load('inputs.json')

results = []
for input in inputs:
    res = build_matrix(**input)
    results.append(res)

json_tricks.dump(results, '.answer.json');
```
