```python
def reconstruct(A, b):
    # U, s, VT = np.linalg.svd(A)
    # results = []
    # res = None
    ### YOUR CODE HERE ###
    U, s, VT = np.linalg.svd(A)
    
    S = np.diag(s)
    
    results = []
    
    for k in range(1, 17):
        S_k = S[:k, :k]
        U_k = U[:, :k]
        VT_k = VT[:k, :]
        
        x_k = np.linalg.inv(S_k) @ U_k.T @ b
        
        x_k_image = x_k.reshape((20, 20))
        results.append(x_k_image)

    return results, U, s, VT
```

```python
inputs = json_tricks.load('.inputs.json')

# print(inputs)

x_rec, U, s, VT = reconstruct(**inputs)

json_tricks.dump(x_rec, '.answer.json')
```
